```python
# Importing all required python libraries required for cleaning.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Accessing the cleaned laptop price dataset
dataset = pd.read_csv('../data/processed/laptop_price_clean.csv')
```

```python
# First view
dataset.head()
```

```python
dataset.info()
```

```python
# Converting ['company','typename', 'screenresolution', 'cpu', 'memory', 'gpu', 'opsys'] columns to category
dataset[['Company', 'TypeName', 'ScreenResolution', 'Cpu', 'Memory', 'Gpu', 'OpSys']] = \
    dataset[['Company', 'TypeName', 'ScreenResolution', 'Cpu', 'Memory', 'Gpu', 'OpSys']].astype('category')

dataset.info()
```

```python
dataset.columns = dataset.columns.str.lower()


plt.figure(figsize=(12,6))
sns.barplot(x='company', y='price_euros', data=dataset)
plt.title('Average Price of Laptops by Company')
plt.xticks(rotation=45)
plt.show()
```

```python
# Barplot of company and price considering the type of laptop
plt.figure(figsize=(12,6))
sns.barplot(x='company', y='price_euros', hue='typename', data=dataset)
plt.title('Average Price of Laptops by Company')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap of all the numerical features
plt.figure(figsize=(12,6))
sns.heatmap(dataset.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Numerical Features')
plt.show()
```

It can be noted that the price is correlating very well with the Ram size.

```python
# Bar plot of the memory size and price
plt.figure(figsize=(12,6))
sns.barplot(x='price_euros', y='memory', data=dataset)
plt.title('Bar Plot of Memory Size vs Price')
plt.show()
```

```python
# Bar plot of screenresolution with price
plt.figure(figsize=(12,6))
sns.barplot(x='screenresolution', y='price_euros', data=dataset)
plt.title('Average Price of Laptops by Screen Resolution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of screenresolution with price considering the type of laptop
plt.figure(figsize=(12,6))
sns.barplot(x='screenresolution', y='price_euros', hue='typename', data=dataset)
plt.title('Average Price of Laptops by Screen Resolution')
plt.xticks(rotation=45)
plt.show()
```
