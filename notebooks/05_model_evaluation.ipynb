```python
# Step 7: Clean and prepare data for machine learning
import re

# Clean RAM column (remove 'GB' and convert to numbers)
df['Ram_GB'] = df['Ram'].str.replace('GB', '').astype(int)
print("RAM column cleaned")

# Clean Weight column (remove 'kg' and convert to numbers)
df['Weight_kg'] = df['Weight'].str.replace('kg', '').astype(float)
print("Weight column cleaned")

# Check the cleaned columns
print(f"\nRAM values: {sorted(df['Ram_GB'].unique())}")
print(f"Weight range: {df['Weight_kg'].min():.2f}kg to {df['Weight_kg'].max():.2f}kg")

# Drop the original text columns
df_clean = df.drop(['Ram', 'Weight'], axis=1)
print("Original text columns removed")

print(f"\nCleaned dataset shape: {df_clean.shape}")
```

```python
# Step 8: Prepare features and target for machine learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Separate features (X) and target (y)
X = df_clean.drop(['Price_euros', 'laptop_ID', 'Product'], axis=1)  # Remove price and ID columns
y = df_clean['Price_euros']  # This is what we want to predict

print("Features (X) columns:")
print(X.columns.tolist())
print(f"\nFeatures shape: {X.shape}")
print(f"Target (prices) shape: {y.shape}")

# Check data types
print(f"\nFeature data types:")
print(X.dtypes)
```

```python
# Step 9: Convert text data to numbers
from sklearn.preprocessing import LabelEncoder

# Make a copy to work with
X_encoded = X.copy()

# List of text columns to encode
text_columns = ['Company', 'TypeName', 'ScreenResolution', 'Cpu', 'Memory', 'Gpu', 'OpSys']

# Create label encoders for each text column
label_encoders = {}

for column in text_columns:
    le = LabelEncoder()
    X_encoded[column] = le.fit_transform(X[column])
    label_encoders[column] = le
    print(f"Encoded {column}: {len(le.classes_)} unique values")

print(f"\nAfter encoding:")
print("All data types:")
print(X_encoded.dtypes)
print(f"\nFinal feature matrix shape: {X_encoded.shape}")
```

```python
# Step 10: Split data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data (80% for training, 20% for testing)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

print("Data split completed:")
print(f"Training set: {X_train.shape[0]} laptops")
print(f"Testing set: {X_test.shape[0]} laptops")
print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")

print(f"\nPrice ranges:")
print(f"Training set prices: €{y_train.min():.2f} to €{y_train.max():.2f}")
print(f"Testing set prices: €{y_test.min():.2f} to €{y_test.max():.2f}")
```

```python
# Step 11: Train a Random Forest model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Create and train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate performance metrics
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print("Model Performance:")
print(f"R-squared (R2) Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): €{mae:.2f}")
print(f"Root Mean Square Error (RMSE): €{rmse:.2f}")
print(f"Average actual price: €{y_test.mean():.2f}")
print(f"Prediction accuracy: {(1 - mae/y_test.mean())*100:.1f}%")
```

```python
# Step 12: Visualize model performance
import matplotlib.pyplot as plt

# Create prediction vs actual plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', linewidth=2)
plt.xlabel('Actual Price (Euros)')
plt.ylabel('Predicted Price (Euros)')
plt.title('Actual vs Predicted Laptop Prices')
plt.grid(True, alpha=0.3)

# Add performance text
plt.text(0.05, 0.95, f'R² = {r2:.3f}\nMAE = €{mae:.0f}',
         transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat'))

plt.tight_layout()
plt.show()

print("Perfect predictions would fall on the red diagonal line.")
print("Points closer to the line = better predictions")
```

```python
# Step 13: Analyze which features matter most for pricing
feature_importance = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': model.feature_importances_
}).sort_values('Importance', ascending=False)

print("Feature Importance Rankings:")
print(feature_importance)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['Feature'], feature_importance['Importance'])
plt.xlabel('Importance Score')
plt.title('Feature Importance for Laptop Price Prediction')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

print(f"\nTop 3 most important features:")
for i, row in feature_importance.head(3).iterrows():
    print(f"{i+1}. {row['Feature']}: {row['Importance']:.3f}")
```

```python
# Step 14: Compare different machine learning models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor

# Initialize different models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42)
}

# Train and evaluate each model
results = {}

for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions
    predictions = model.predict(X_test)

    # Calculate metrics
    r2 = r2_score(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))

    results[name] = {
        'R2_Score': r2,
        'MAE': mae,
        'RMSE': rmse,
        'Accuracy': (1 - mae/y_test.mean()) * 100
    }

    print(f"{name}:")
    print(f"  R2 Score: {r2:.4f}")
    print(f"  MAE: €{mae:.2f}")
    print(f"  Accuracy: {(1 - mae/y_test.mean()) * 100:.1f}%")
    print()

# Find best model
best_model = max(results.keys(), key=lambda x: results[x]['R2_Score'])
print(f"Best performing model: {best_model}")
print(f"Best R2 Score: {results[best_model]['R2_Score']:.4f}")
```

```python
# Step 15: Create prediction function for new laptops
def predict_laptop_price(company, type_name, inches, ram_gb, weight_kg,
                        screen_res, cpu, memory, gpu, os):
    """
    Predict laptop price based on specifications
    """
    # Create a sample with the input specifications
    sample_data = pd.DataFrame({
        'Company': [company],
        'TypeName': [type_name],
        'Inches': [inches],
        'ScreenResolution': [screen_res],
        'Cpu': [cpu],
        'Memory': [memory],
        'Gpu': [gpu],
        'OpSys': [os],
        'Ram_GB': [ram_gb],
        'Weight_kg': [weight_kg]
    })

    # Encode the categorical variables using our trained encoders
    sample_encoded = sample_data.copy()
    for column in text_columns:
        try:
            sample_encoded[column] = label_encoders[column].transform([sample_data[column].iloc[0]])
        except ValueError:
            # If new category not seen in training, use most common category
            sample_encoded[column] = [0]  # Use first encoded value as fallback

    # Make prediction using the best model
    prediction = models[best_model].predict(sample_encoded)[0]
    return prediction

# Example prediction
print("Example Laptop Price Prediction:")
print("Specifications:")
print("- Company: Dell")
print("- Type: Gaming")
print("- Screen: 15.6 inches")
print("- RAM: 16GB")
print("- Weight: 2.5kg")

# Make prediction (using common values from the dataset)
try:
    predicted_price = predict_laptop_price(
        company='Dell',
        type_name='Gaming',
        inches=15.6,
        ram_gb=16,
        weight_kg=2.5,
        screen_res='1920x1080',  # Common resolution
        cpu='Intel Core i7',     # Common CPU
        memory='1TB HDD',        # Common memory
        gpu='NVIDIA GeForce',    # Common GPU
        os='Windows 10'          # Common OS
    )
    print(f"\nPredicted Price: €{predicted_price:.2f}")
except:
    print("\nNote: Adjust the example specifications to match your dataset values")
    print("Check unique values in each column to use valid inputs")
```
